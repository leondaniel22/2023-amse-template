### 4.3. Correalation between Dataset 1 and Dataset 2

Merge the dataframes of dataset 1 and dataset 2 into a new dataframe. Therefore an outer join on column "Date" is used. 

```python
df3 = pd.merge(df1, df2, on='Date', how='outer')
```

Show the 10 first rows of the new dataframe

```python
df3.head(10)
```

Group the new dataframe by months and use the sum aggregation for all columns of dataset 1 and the mean aggregation for all columns of dataset 2

```python
aggs = {}
for index in df1.columns:
    if not index == 'Date':
        aggs[index] = 'sum'
for index in df2.columns:
    if not index == 'Date':
        aggs[index] = 'mean'
dg = df3.groupby(pd.Grouper(key='Date', freq='M')).agg(aggs) # Group by month
dg.index = dg.index.strftime('%B') # Use name of manth as new Date value
dg
```

Sort the new montly grouped dataframe by the amount of total bikers in descending order

```python
dg.sort_values(['Total bikers'], ascending=[False])
```

Showing the amount of bikers for different average temperature ranges

```python
import numpy as np
d = df3[df1.columns]
d.groupby(pd.cut(df3["Average temperature"], np.arange(-20.0, 50.0, 10.0))).sum(numeric_only=True)
```

Showing the amount of bikers for different total rainfall ranges 

```python
d = df3[df1.columns]
d.groupby(pd.cut(df3["Total rainfall"], np.arange(0, 70.0, 5.0))).sum(numeric_only=True)
```

Showing the different average weather conditions for total biker ranges

```python
d = df3[df2.columns]
d.groupby(pd.cut(df3["Total bikers"], np.arange(0, 15779+1500, 1500))).mean(numeric_only=True)
```
